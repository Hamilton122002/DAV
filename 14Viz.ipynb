```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("cereal.csv")
df
```

```python
plt.bar(df['mfr'],df['calories'])
plt.title("Calories Bar Chart")
plt.show
```

```python
sb.lineplot(x='sodium',y='mfr', data=df)
plt.show()
```

```python
plt.scatter(df['sugars'] ,df['carbo'])
plt.show
```

```python
sb.lineplot(x='mfr', y='sodium', data=df)
plt.show
```

```python
plt.plot(df['sugars'])
plt.plot(df['fiber'])
plt.ylabel('Quantity')
plt.show
```

```python
plt.hist(df['potass'])
plt.title('Histogram')
plt.show()
```

```python
sb.barplot(x='mfr', y='rating', data= df, hue='vitamins')
plt.show()
```

```python
from bokeh.plotting import figure, output_file, show
graph = figure(title = "Bokeh Scatter Plot")
graph.scatter(df['sodium'], df['fiber'])
show(graph)
```

```python
fig = plt.figure(figsize = (10,7))
ax = plt.axes(projection = "3d")
ax.scatter3D(df['sodium'], df['potass'], df['sugars'])
plt.title("3d Scatter Plot")
plt.show()
```

```python
sb.set(style='ticks', color_codes = True)
iris = sb.load_dataset("iris")
g = sb.pairplot(iris)
plt.show()
```

```python
my_data = [300, 500, 700]
my_labels = 'Tasks Pending', 'Tasks Ongoing', 'Tasks Completed'
plt.pie(my_data, labels=my_labels, autopct='%1.1f%%')
plt.title('My Tasks')
plt.axis('equal')
plt.show()
```
